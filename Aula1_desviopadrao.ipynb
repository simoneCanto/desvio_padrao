```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


```

```python
# Dados extraídos da planilha
dias = np.arange(1, 11)
medias_diarias = np.array([10.06, 10.08, 10.3, 10.2, 10.22, 9.84, 10.3, 9.8, 10.16, 9.84])
media_geral = 10.06
limite_superior = 10.47
limite_inferior = 9.65

```

```python
# Criando o gráfico de controle
plt.figure(figsize=(10, 5))
plt.plot(dias, medias_diarias, marker='o', linestyle='-', label='Média Diária')
plt.axhline(media_geral, color='green', linestyle='--', label='Média Geral (X̄)')
plt.axhline(limite_superior, color='red', linestyle='--', label='LCS (Limite Superior)')
plt.axhline(limite_inferior, color='red', linestyle='--', label='LCI (Limite Inferior)')

# Configurações do gráfico
plt.xticks(dias)
plt.xlabel('Dias de Produção')
plt.ylabel('Média do Diâmetro dos Parafusos')
plt.title('Gráfico de Controle X̄')
plt.legend()
plt.grid(True)

# Exibir o gráfico
plt.show()
```
